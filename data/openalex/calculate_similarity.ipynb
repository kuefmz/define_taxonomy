```python
import spacy
import json
```

```python
def preprocess_term(term):
	# Split the term into words and filter out numeric-only words
	filtered_words = [word for word in term.split(' ') if not word.isnumeric()]
	# Rejoin the words to form the preprocessed term
	preprocessed_term = ' '.join(filtered_words)
	return preprocessed_term
```

```python
def compute_similarity(terms):
    nlp = spacy.load("en_core_web_md")
    doc_terms = [nlp(term) for term in terms]
    
    print("Semantic similarity among AI terms:")
    for i, doc1 in enumerate(doc_terms):
        for j, doc2 in enumerate(doc_terms):
            if i < j:
                similarity = doc1.similarity(doc2)
                print(f"Similarity between '{doc1}' and '{doc2}': {similarity:.4f}")
```

```python
def calculate_similarity(json_data, threshold):
    output_json = {}
    nlp = spacy.load("en_core_web_md")
    
    cnt = 1
    for title, data in json_data.items():
        print(f'Counter: {cnt}')
        cnt += 1
        openaire_terms = data['openaire'].split(' | ')
        openalex_terms = data['openalex'].split(' | ')
        
        # Initialize output structure for the current title
        output_json[title] = {
            'openaire': data['openaire'],
            'openalex': data['openalex'],
            'similarity': [],
            'matches': [],
        }
        
        # Calculate similarity between each pair of terms
        for openaire_term in openaire_terms:
            for openalex_term in openalex_terms:
                openaire_term = preprocess_term(openaire_term)
                openalex_term = preprocess_term(openalex_term)
                doc1 = nlp(openaire_term)
                doc2 = nlp(openalex_term)
                similarity = doc1.similarity(doc2)
                
                # Add the similarity result to the output JSON
                output_json[title]['similarity'].append({
                    'openaire': openaire_term,
                    'openalex': openalex_term,
                    'similarity': similarity
                })
                #print(output_json[title]['similarity'])
                
                if similarity > threshold:
                    output_json[title]['matches'].append({
                        'openaire': openaire_term,
                        'openalex': openalex_term,
                        'similarity': similarity
                    })
    with open("data_100_similarities_2.json", 'w') as file:
        json.dump(output_json, file, indent=4)
    
    return output_json
```

```python
out_json = None
with open("data_100.json", 'r') as file:
    data = json.load(file)
    out_json = calculate_similarity(data, 0.5)
```
