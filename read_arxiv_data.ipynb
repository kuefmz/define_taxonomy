```python
import json
```

```python
categories = {}
titles = []
with open('arxiv.json', 'r') as f:
    cnt = 0
    while True:
        line = f.readline()
        if not line:
            break
        item = json.loads(line)
        #print(item.keys())
        cat = item['categories']
        titles.append(item['title'])
        
        if cat not in categories.keys():
            categories[cat] = 1
        else:
            categories[cat] += 1
        cnt += 1
```

```python
print(categories.keys())
```

```python
cnt = 0
for t in titles:
    cnt += 1
    if cnt > 100:
        break
    print(' - ', t)
```
